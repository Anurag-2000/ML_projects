```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('Car details v3.1.csv')
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df.drop('name', axis=1,inplace=True)
df.head(5)
```

```python
df['mileage']=df['mileage'].str.replace('kmpl','')
df['mileage']=df['mileage'].str.replace('km/kg','').astype(float)
df.head(5)
```

```python
df['engine']=df['engine'].str.replace('CC','').astype(float)
df.head(5)
```

```python
df['max_power']=df['max_power'].str.replace('bhp','')
df.head()
df['max_power']=df['max_power'].astype(float)
```

```python
df.dtypes
```

```python
print(df['fuel'].unique())
print(df['seller_type'].unique())
print(df['transmission'].unique())
print(df['owner'].unique())
```

```python
df['Total_years']= 2021 - df['year']
df.head(10)
```

```python
df.drop('year',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.drop('torque',axis=1,inplace=True)
df.head()
```

```python
final_df=pd.get_dummies(df,drop_first=True)
```

```python
final_df.head()
```

```python
final_df.shape
```

```python
final_df.describe()
```

```python
final_df.corr()
```

```python
corrmat=final_df.corr()
top_corr_features=corrmat.index
plt.figure(figsize=(20,20))
#plot heat map
g=sns.heatmap(final_df[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

```python
X=final_df.iloc[:,1:]
y=final_df.iloc[:,0]
```

```python
X.head()
```

```python
y.head()
```

```python
#print(final_df.replace(r'^\s*$', np.nan, regex=True))
X.head(15)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X.iloc[:, 1:5])
X.iloc[:, 1:5] = imputer.transform(X.iloc[:, 1:5])
```

```python
X.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
X_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 42)
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
# print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))

```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
z=list(y_test)
print("Prediction|Actuall")
for i in range(0,y_pred.shape[0]):
    print("{:.2f}  {}".format(y_pred[i],z[i]))
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = regressor, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```
